```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings

warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
```

```python
''' reading dataset '''
df = pd.read_csv('haberman.csv')
```

```python
''' displaying first 5 rows of data ''' 
df.head()
```

```python
''' renaming columns '''
df.columns = ['age', 'year', 'nodes', 'status']
```

```python
df.head()
```

```python
''' shape of data '''
df.shape
```

```python
''' checking null values in data ''' 
df.isnull().sum()
```

```python
df.head()
```

```python
''' value counts of status '''
status = df.status.value_counts()

''' plotting '''
plt.figure(figsize=(8, 5))
sns.barplot(status.index, status)
plt.xlabel('Status', fontsize=15);
plt.ylabel('Count', fontsize=15);
```

```python
''' scatter plot of age and nodes '''
plt.figure(figsize=(10, 5))
sns.scatterplot(x='age', y='nodes', data=df)
plt.xlabel('age', fontsize=15)
plt.ylabel('nodes', fontsize=15);
```

```python
''' correlation '''
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True)
plt.title('Correlation Matrix', fontsize=20);
```

```python
''' Multivariate Analysis '''
plt.figure(figsize=(20, 8))
sns.pairplot(df, hue='status');
```

```python
''' distribution plot '''
for i, column in enumerate(df.columns):
    sns.distplot(df[column])
    plt.title(column, fontsize=20);
    plt.show()
```

```python
''' boxplot '''
for column in df.columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(df[column])
    plt.title(column, fontsize=20);
```

```python
''' histogram plot '''
for column in df.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df[column], kde=True)
    plt.title('Histogram plot of {}'.format(column), fontsize=20);
```

```python
''' box plot of status and nodes '''
plt.figure(figsize=(10, 5))
sns.boxplot(x='status',y='nodes',data=df)
plt.title('Box plot', fontsize=20);
plt.xlabel('status', fontsize=15)
plt.ylabel('nodes', fontsize=15)
plt.show();
```

```python
''' violin plot of stauts and node '''
plt.figure(figsize=(10, 5))
sns.violinplot(x='status',y='nodes',data=df)
plt.title('Violin plot', fontsize=20)
plt.xlabel('status', fontsize=15);
plt.ylabel('nodes', fontsize=15);
```

```python
''' univariate analysis of status and nodes '''
plt.figure(figsize=(10, 5))
sns.jointplot(x='status', y='nodes', data=df, kind='kde')
plt.title('Univariate Analysis')
plt.xlabel('staus', fontsize=15)
plt.ylabel('nodes', fontsize=15);
```
