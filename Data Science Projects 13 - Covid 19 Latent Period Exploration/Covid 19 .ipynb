# Abstract

The incubation period for the COVID-19 is reported. We can't say the same for the latent period (the period since the person is infected but does not spread the infection yet), which is vital characteristic of every infectious disease. Knowing the latent period gives an ability to run simulations which are far more accurate. In this study I try to estimate the latent period for COVID-19 caused by SARS-CoV-2 virus via computational simulations using SEIR model with different values controlling the latent period length (e.g. latent period parameter sweep). I try to fit all other parameters to match the observed statistics the best. I do such modelling for 5 different locations worldwide which are currently experiencing different stages of the epidemic.

As a result, Europe locations suggest that latent period is between 2 and 4 days: Austria case suggests the value of 48-96 hours, Italy case suggests the value greater than 48 hours. While China cases suggest shorter latent period.

# Introduction

## The data

I use the time series data (infected count, recovered count and deaths count) published by the [John Hopkins University](https://github.com/CSSEGISandData/COVID-19) as ground truth observations.

I experiment with these 5 locations:
* Austria (infection peak has passed)
* Beijing, China (there were 2 peaks)
* Heilongjiang, China (two more distinct peaks)
* Italy (Peak is approaching)
* South Korea (single peak passed long ago)


## Latent period estimation

Latent period can be estimated as reciprocal to sigma parameter. Thus `latent period = 1.0 / sigma`

## The parameter fitting procedure

I let the optimizer to fit the following parameters:
* Infection start date
* number of the exposed people count on the first day of infection
* number of the infected people count on the first day of infection
* beta (separate value per week) - infection rate
* gamma - recovery rate


```r
require(ggplot2)
require(tidyr)


options(repr.plot.width=15, repr.plot.height=8)
```

```r
# loading the results of the computational experiment
df1 <- read.csv('../input/covid19-latent-period-computational-experiment/fittedParams.csv')
```

```r
# suppling parameters dataset with max infected count (needed for loss value normalization between locations)
uniqueLocationsDf <- df1[!duplicated(df1$tableFileKey),]
N <- nrow(uniqueLocationsDf)
maxInfectedObeservedDf <- NULL
for(i in 1:N) {
    row <- uniqueLocationsDf[i,]
    tsDf <- read.csv(file.path('../input/covid19-latent-period-computational-experiment/simTables',paste0(row$tableFileKey,'.csv')))
    maxInfectedObserved <- max(tsDf$infected.obs)
    rowDf <- data.frame(tableFileKey=row$tableFileKey, infectedMaxObserved = maxInfectedObserved)
    if(is.null(maxInfectedObeservedDf))
        maxInfectedObeservedDf <- rowDf
    else
        maxInfectedObeservedDf <- rbind(maxInfectedObeservedDf,rowDf)
}

df1 <- merge(df1,maxInfectedObeservedDf, by='tableFileKey')
#df1
```

```r
# adding some additional columns

# modeled location full name
df1$Location <- as.character(df1$Country)
provinceSpecified <- nchar(as.character(df1$Province))>0
df1$Location[provinceSpecified] <- paste0(df1$Province[provinceSpecified],' - ',df1$Country[provinceSpecified])
df1$Location <- as.factor(df1$Location)

df1$LossNormalized <- df1$Loss / df1$infectedMaxObserved 
# FitGoodness = Loss reciprocial : higher is better
df1$NormalizedFitGoodness <- 1.0 / df1$LossNormalized

# LatentPeriod, which is treated as factor
df1$LatentPeriod_hours <- as.factor(round(24.0 / df1$Sigma)) # in hours
```

```r
# Utility functions

loadSimulationsTS <- function(location) {
    locSpecDf <- df1[df1$Location == location,]
    sigmaAndFilenameDf <- locSpecDf[,c("LatentPeriod_hours","tableFileKey")]
    rowCount <- nrow(sigmaAndFilenameDf)
    
    #print(sigmaAndFilenameDf)
    acc <- NULL
    
    for(rowNum in (1:rowCount)) {
        latentPeriod <- sigmaAndFilenameDf[rowNum,"LatentPeriod_hours"]
        tsTableFile <- sigmaAndFilenameDf[rowNum,"tableFileKey"]
        tsDf <- read.csv(file.path('../input/covid19-latent-period-computational-experiment/simTables',paste0(tsTableFile,'.csv')))
        tsDf <- tsDf[,c(-3,-4)] # removing obs columns
        tsDf$LatentPeriod_hours <- latentPeriod
        if(is.null(acc))
            acc <- tsDf
        else
            acc <- rbind(acc,tsDf)
    }
    
    acc$Date <- as.Date(acc$Date)
    # renaming columns
    acc$exposed <- acc$exposed.pred
    acc$infected <- acc$infected.pred
    acc$exposed.pred <- NULL
    acc$infected.pred <- NULL
    # 3 timeseries columns into a key-value columns to be able to plot with ggplot
    acc <- gather(acc, key="TsType",value="People",exposed, infected)
    acc$TsType <- as.factor(acc$TsType)
    #print(summary(acc))
    acc
}

loadObservationsTs <-  function(location) {
    tsTableFile <- df1[df1$Location == location,"tableFileKey"][1]
    tsDf <- read.csv(file.path('../input/covid19-latent-period-computational-experiment/simTables',paste0(tsTableFile,'.csv')))
    tsDf <- tsDf[,1:4] # dates and obs columns
    tsDf$Date <- as.Date(tsDf$Date)
    names(tsDf)[3:4] <- c("infected","removed")
    tsDf <- gather(tsDf, key="TsType",value="People",infected)
    tsDf$TsType <- as.factor(tsDf$TsType)
    tsDf
}

figure1 <- function(locationOfInterest) {
    locationDf <- df1[df1$Location == locationOfInterest,]

    plot(ggplot(locationDf, aes(x=LatentPeriod_hours, y=Loss)) +
        geom_bar(stat="identity") +
        ggtitle(paste0("Achieved loss for different supposed latent periods for ",locationOfInterest)))
}

figure2 <- function(locationOfInterest, top = 5) {
    locationDf <- df1[df1$Location == locationOfInterest,]
    
    # Considering top N parameter fits
    locationDf <- locationDf[order(locationDf$Loss)[1:top],]

    # Taking top N timeseries and joining
    locationPredTsDf <- loadSimulationsTS(locationOfInterest)
    locationPredTsDf <- locationPredTsDf[locationPredTsDf$LatentPeriod_hours %in% unique(locationDf$LatentPeriod_hours), ]
    locationPredTsDf <- merge(locationPredTsDf, locationDf, by='LatentPeriod_hours')
    #print(beijingPredTsDf)

    locationObsTsDf <- loadObservationsTs(locationOfInterest)

    linetypes <- c("infected"='solid','exposed'='dashed')
    shapes <- c("infected"=3)
    cbPalette <- c("#000000", "#E69F00", "#56B4E9", "#009E73", "#F0E442", "#0072B2", "#D55E00", "#CC79A7")

    ggplot() +
        theme_bw() +
        scale_linetype_manual(values=linetypes,name="Model output") +
        scale_shape_manual(values=shapes,name="Observed") +
        scale_colour_manual(values=cbPalette,name=" Latent Period (hours)") +
        ggtitle(paste0("Top ",top," model fits for ",locationOfInterest)) +
        geom_line(data=locationPredTsDf, aes(x=Date, y=People, color=LatentPeriod_hours, linetype=TsType), size=1) +
        geom_point(data=locationObsTsDf, aes(x=Date, y=People, shape=TsType))
        

}
```

# Italy case

```r
# Exploring Italy
figure1('Italy')
```

We can see that short latent period values (2 hours - 48 hours) exhibit higher values of achieved loss value, thus less likely to describe the observations in a valid way.

But we can't see any explicit single latent period value longer than 48 hours, that exhibits much lower loss. All of values starting from 72 hours and higher seem to be possible.

```r
figure2("Italy", top = 8)
```

Top fits match the observed infected count almost perfectly for Italy. Wide range of "valid" latent periods are equaly possible.
