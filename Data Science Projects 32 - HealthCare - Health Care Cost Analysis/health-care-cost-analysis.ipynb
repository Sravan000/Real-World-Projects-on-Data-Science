```r
mydata <- read.csv("../input/healthcare-cost/HospitalCosts.csv")
hospcost = copy(mydata)
```

```r
#Primary Data Inspection
summary(hospcost)
str(hospcost)
sum(is.na(hospcost))
dim(hospcost)
hospcost <- na.omit(hospcost)
summary(hospcost)
dim(hospcost)
```

```r
#Creating New Variable for categorization of age group.

hospcost["AgeGroup"] <-ifelse(between(hospcost$AGE, 0,5),"0-5",
                              ifelse(between(hospcost$AGE, 6,10),"6-10",
                              ifelse(between(hospcost$AGE, 11,15),"11-15","16-20")))
str(hospcost)
```

```r
#Importing Libraries
library(dplyr)
library(ggplot2)
library(gridExtra)
options(scipen = 999)
```

```r
#Task1: To record the patient statistics, the agency wants to find the age category of people who frequently visit the hospital and has the maximum expenditure.

LOS <- hospcost%>%
  group_by(AgeGroup)%>%
  summarise(LOS = sum(LOS))%>%
  ggplot(aes(AgeGroup,LOS,fill=AgeGroup))+
  geom_bar(stat='identity', width = .8)+
  geom_text(aes(label=LOS),vjust=1.0,color="white",size=3.0)+
  ggtitle("AgeGroup Wise Tenure of Patients")+
  xlab("Age Group") + ylab('Length of Stay (Days)')+
  theme_classic()

TotCost <- hospcost%>%
  group_by(AgeGroup)%>%
  summarise(TotCost = sum(TOTCHG))%>%
  ggplot(aes(AgeGroup,TotCost,fill=AgeGroup))+
  geom_bar(stat='identity', width = .8)+
  geom_text(aes(label=TotCost),vjust=1.0,color="white",size=3.0)+
  ggtitle("AgeGroup Wise Total Cost of Patients")+
  xlab("Age Group") + ylab('Total Cost in Doller')+
  theme_classic()

## Share % Of Cost and Stay


LOS1 <- hospcost%>%
  group_by(AgeGroup)%>%
  summarise(LOSP = sum(LOS))%>%
  ggplot(aes("",LOSP,fill=AgeGroup))+
  geom_bar(stat='identity', width = .8)+
  coord_polar("y", start=0)+
  geom_text(aes(label=round(LOSP/sum(LOSP)*100,2)),vjust=1.0,color="white",size=3.0)+
  ggtitle("AgeGroup Wise Stack %age of Stay")+
  theme_void()

TotCost1 <- hospcost%>%
  group_by(AgeGroup)%>%
  summarise(TotCostP = sum(TOTCHG))%>%
  ggplot(aes("",TotCostP,fill=AgeGroup))+
  geom_bar(stat='identity', width = .8)+
  coord_polar("y", start=0)+
  geom_text(aes(label=round(TotCostP/sum(TotCostP)*100,2)),vjust=1.0,color="white",size=3.0)+
  ggtitle("AgeGroup Wise Stack %age of Expense")+
  theme_void()

grid.arrange(LOS,TotCost,LOS1,TotCost1, nrow=2)

## the Age Group of 0-5 (Basically Children) has maximum Hospital Visit and Expenses. However age group 6-10 has minimum Visits and Expenses.
```

```r
#Task2: In order of severity of the diagnosis and treatments and to find out the expensive treatments, the agency wants to find the diagnosis-related group that has maximum hospitalization and expenditure.
#Chaning Diagnosis Group int Factor.

hospcost$APRDRG <- factor(hospcost$APRDRG)
str(hospcost)

##Remove those Diagnosis group which has less than 10 days stay because these are not more significant as well as visualization is not being intutive.
LOSD <- hospcost%>%
  group_by(APRDRG)%>%
  summarise(LOSD = sum(LOS))%>%
  filter(LOSD > 5)%>%
  ggplot(aes(APRDRG,LOSD))+
  geom_bar(stat='identity', color='green',width = .95)+
  geom_text(aes(label=APRDRG),vjust=.01,color="black",size=3.0)+
  ggtitle("DiagnosisGroup Wise Length of Stay of Patients")+
  xlab("Diagnosis Group") + ylab('Length of Stay (Days)')+
  theme_classic()

##Remove those Diagnosis group which has less than 20000$ cost because these are not more significant as well as visualization is not being intutive.
TotCostD <- hospcost%>%
  group_by(APRDRG)%>%
  summarise(TotCostD = sum(TOTCHG))%>%
  filter(TotCostD > 20000)%>%
  ggplot(aes(APRDRG,TotCostD))+
  geom_bar(stat='identity',color='green', width = .8)+
  geom_text(aes(label=APRDRG),vjust=.01,color="black",size=3.0)+
  ggtitle("DiagnosisGroup Wise  Total Cost of Patients")+
  xlab("Diagnosis Group") + ylab('Total Cost in Doller')+
  theme_classic()

grid.arrange(LOSD,TotCostD, nrow=1)

## By looking the data Diagnosis Group 640 has maximum hospitalization and expenditure.
```

```r
#Task3: To make sure that there is no malpractice, the agency needs to analyze if the race of the patient is related to the hospitalization costs.

library(car)
#Scatter Plot
scatterplot(TOTCHG~RACE, hospcost)

#Correlation Test
RaceInflCor <- cor.test(hospcost$RACE,hospcost$TOTCHG,method='pearson')
RaceInflCor

#Linear Regression Test
RaceInflLM <- lm(TOTCHG~RACE, hospcost)
summary(RaceInflLM)

## by testing with Correlation Test and Linear Regression Test, Race is not related to hospitalization cost.
```

```r
#Task4: To properly utilize the costs, the agency has to analyze the severity of the hospital costs by age and gender for the proper allocation of resources. 
## Testing with ANOVA

costaov <- aov(TOTCHG~AGE+FEMALE,hospcost)
summary(costaov)

## Testing with Linear Regression

costlm <- lm(TOTCHG~AGE+FEMALE,hospcost)
summary(costlm)


## Both Age and Female has significant effect on Hospital Cost looking by P Value which is less than 0.05.
```

```r
#Task5: Since the length of stay is the crucial factor for inpatients, the agency wants to find if the length of stay can be predicted from age, gender, and race.

##Test with Linear Regression

StateLM <- lm(LOS~AGE+FEMALE+RACE, hospcost)
summary(StateLM)

#Since P value is higher than 0.05, so we can say that Age, Gender and Race not impact Length of Stay, so these can not be predictor for Lengty of Stay.

```

```r
#Task6: To perform a complete analysis, the agency wants to find the variable that mainly affects hospital costs.
#Test with Linear Regression
#Change the data type of Diagnosis Group factor to numeric for model.

hospcost$APRDRG <- as.numeric(hospcost$APRDRG)
str(hospcost)
#Linear Test
StateLM <- lm(TOTCHG~.,hospcost[-7])
summary(StateLM)

## By looking the result Age, Length of Stay and Diagnosis Group affect the Hospital Cost.

##End of Project##

```
